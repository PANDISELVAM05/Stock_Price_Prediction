```python
#DATA CLEANING SCRIPT
import pandas as pd
import numpy as np
import re
import sys
```

```python
df=pd.read_csv('D:\Gradtwin\Stock Price Prediction\Oil_Data.csv')
df.head(10)
```

```python
print(df.columns)
```

```python
print(df.dtypes)
```

```python
print(df.info())
print(df.isnull().sum())
```

```python
def contains_only_special_chars(value):
    if pd.isna(value):
     
     
        return True
    return bool(re.fullmatch(r'[^a-zA-Z0-9]+', str(value)))

df_cleaned = df[~df.applymap(contains_only_special_chars).any(axis=1)]
```

```python
print(df_cleaned.head())
```

```python
cleaned_file_path = r"D:\Gradtwin\Stock Price Prediction\Oil_Data_Cleaned.csv"
df_cleaned.to_csv(cleaned_file_path, index=False)
print(f"\n\n\nCleaned data saved to {cleaned_file_path}")
```
